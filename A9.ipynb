```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('Titanic.csv')
df.head()
```

```python
df=df.set_index('PassengerId')
```

```python
df['SibSp'].plot(kind='hist')
plt.title('Histogram of Number of Siblings on Board')
plt.xlabel('Number of Siblings')
```

```python
df['Parch'].plot(kind='hist')
plt.title('Histogram of Number of Parents on Board')
plt.xlabel('Number of parents')
```

```python
df['Fare'].plot(kind='line')
plt.title('Fare Paid')
plt.ylabel('Price in British Pound')
```

```python
df['Fare'].plot(kind='hist')
plt.xlabel('Price in British Pound')
plt.title('Histogram of Fares')
```

```python
df_embark=df.set_index('Embarked')
df_embark
```

```python
df_ecount=df.groupby('Embarked').count()
```

```python
df_ecount['Name'].plot(kind='pie')
plt.title('Percentage of Embarks From Each Port')
plt.ylabel('')
```

```python
df_eavg=df_embark['Fare'].groupby('Embarked').mean()
df_eavg
```

```python
df_eavg.plot(kind='bar')
plt.title('Average Price of Ticket From Each Port')
plt.ylabel('Price in British Pound')
```

```python
df_sex=df.set_index('Sex')
df_sex
```

```python
df_scount=df.groupby('Sex').count()
df_scount
```

```python
df_scount['Name'].plot(kind='pie')
plt.title('Sex Distribution')
plt.ylabel('')
```
